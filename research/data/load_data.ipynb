```python
import requests
from tqdm import tqdm
import cv2

def load_images(df_, urls_, path):
    unsuccessful = []
    for i in tqdm(df_.index):
        response = requests.get(urls_.loc[i], headers={'User-Agent': "Magic Browser"}, stream=True)
        if response.status_code != 200:
            unsuccessful.append(i)
            continue
        img_data = response.content
        with open(f'{path}/image_{i}.jpg', 'wb') as handler:
            handler.write(img_data)

        # test if image is readable
        if cv2.imread(f'{path}/image_{i}.jpg') is None:
            unsuccessful.append(i)
    
    range_indexes = list(set(df_.index) - set(unsuccessful))
    range_indexes.sort()
    
    df_ = df_.loc[range_indexes]
    
    return df_
```

```python
new_df_train = load_images(df_train, image_urls_train, 'images')
```

### test

```python
import torchvision.datasets as dset

coco_train = dset.CocoDetection(root = "coco_dataset/train2014",
                                annFile = "coco_dataset/annotations/captions_train2014.json")

coco_val = dset.CocoDetection(root = "coco_dataset/val2014",
                                annFile = "coco_dataset/annotations/captions_val2014.json")
```

```python
import json
with open('coco_dataset/captions_ru.json', 'r') as f:
    coco_captions = json.load(f)
```

```python
def find_by_id(im_id):
    ans = []
    for i in coco_captions:
        if i['image_id'] == im_id:
            ans.append(i)
    return ans
```

```python
from tqdm import tqdm
sorted_captions_train = []
for i in tqdm(coco_train):
    sorted_captions_train.append(find_by_id(i[1][0]['image_id']))
```

```python
with open('coco_dataset/captions_ru_train.json', 'w+') as f:
    json.dump(sorted_captions_train, f)
```

```python
from tqdm import tqdm
sorted_captions_val = []
for i in tqdm(coco_val):
    sorted_captions_val.append(find_by_id(i[1][0]['image_id']))
```

```python
with open('coco_dataset/captions_ru_val.json', 'w+') as f:
    json.dump(sorted_captions_val, f)
```
